```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes1 = fig.add_axes([0.2, 0.5, 0.25, 0.25])

axes.set_xlim(0,8)
axes.set_ylim(0, 8000)
axes.set_xlabel('A')
axes.set_ylabel('B')
axes.set_title('Power of 4')

axes1.plot(a,b)

axes1.set_xlim(1,2)
axes1.set_ylim(0, 50)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Zoomed In')

axes.plot(a,b)
```

```python
#figure parameters and saving fig

fig = plt.figure(figsize=(2,2), dpi=150)

axes = fig.add_axes([0,0,1,1])

axes.plot(a,b)

fig.savefig('new_figure.png', bbox_inches='tight')
```
