```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('String Title')
```

```python
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0,6)
plt.ylim(0,15)
plt.title('String Title')
```

```python
#help(plt.savefig)
```

```python
# to save as png

plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0,6)
plt.ylim(0,15)
plt.title('String Title')
plt.savefig('myfirstplot.png')
```
