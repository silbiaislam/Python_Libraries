```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 11)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x, label='X vs X')
ax.plot(x, x**2, label='X vs X^2')
ax.legend(loc=(1.1, 0.5))
```

```python
n = np.linspace(1, 5)
```

```python
from scipy.special import factorial

fig = plt.figure(dpi=125)
ax = fig.add_axes(rect=[0, 0, 1, 1])
ax.set_title("Big-O Complexity")
ax.axhline(label='O(1)')
ax.plot(n, n, label='O(n)')
ax.plot(n, np.log(n), label='O(log n)')
ax.plot(n, n * np.log(n), label='O(n log n)')
ax.plot(n, n**2, label='O(n^2)')
ax.plot(n, 2**n, label='O(2^n)')
ax.plot(n, factorial(n), label='O(n!)')
ax.legend()
fig.savefig('Big-O Chart.png')
```
