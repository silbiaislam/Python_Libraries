## Method Calls

```python
def last_four(num):
    return str(num)[-4:]
```

```python
last_four(123456789)
```

```python
df['time'].apply(last_four)
```

```python
df['total_bill'].mean()
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
```

```python
df['yelp']
```

```python
df
```

```python
#apply on multiple columns 

df['total_bill'].apply(lambda num: num*2)
```

```python
def quality(total_bill,tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
quality(16.99,1.01)
```

```python
df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
```

```python
df['Quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
```

```python
df
```

```python
# np.vectorize, performs faster
# The purpose of np.vectorize is to transform funtions which are not numpy-aware.

df['Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
```
